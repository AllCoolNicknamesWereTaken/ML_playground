```
import tensorflow as tf
import numpy as np
import pandas as pd
import plotly.graph_objects as go
tf.__version__
```

```
function = lambda w: w ** 2 - 4 * w
```

```
derivative = lambda w: 2 * w - 4 
```

```
w = np.arange(-10, 10, 0.1)
```

```
loss = function(w)
w_0 = -8

tangent_line = lambda w: derivative(w_0) * (w - w_0) + function(w_0)
tangent_line_values = tangent_line(w)
```

```
df = pd.DataFrame({'w' : w, 'loss' : loss, 'tangent_line' : tangent_line_values})
df.head()
```

```
fig = go.Figure(data=[go.Scatter(x=df['w'], y=df['loss'], name='function'),
                      go.Scatter(x=df['w'], y=df['tangent_line'], name='tangent line'),
                      go.Scatter(x=[w_0], y=[function(w_0)], marker_size=10, name='piont')],
                layout=go.Layout(width=800, title='Loss function'))
fig.show()
```

```
def gradient_descent(derivative, learning_rate = 0.001, w_0 = -1, max_iters = 10000, precision = 0.0000001, epoch = False):

  points = []
  prev_step = 1
  iters = 0

  while precision < prev_step and iters < max_iters:
    prev_w_0 = w_0
    w_0 = w_0 - learning_rate * derivative(w_0)
    prev_step = abs(w_0 - prev_w_0)
    iters += 1
    points.append(w_0)
    if epoch == True:
      print(f'iteracja: # {iters}, puntkt: {w_0}')
```

```
 function_2 = lambda w: (w + 8) * (w - 5) * (w - 10) * (w + 5) * (w + 5) * (w - 2)
 from sympy import Symbol, lambdify

 w = Symbol('w')
 f = (w + 8) * (w - 5) * (w - 10) * (w + 5) * (w + 5) * (w - 2)
 f_diff = f.diff(w)
 derivative_2 = lambdify(w,  f_diff)
 del w
```

```
w = np.arange(-10, 10, 0.1)
loss = function_2(w)
w_0 = -4

target_line = lambda w: derivative_2(w_0) * (w - w_0) + function_2(w_0)
target_line_values = target_line(w)

df = pd.DataFrame({'w': w, 'loss': loss, 'target_line': target_line_values})

fig = go.Figure(data=[go.Scatter(x=df['w'], y=df['loss'], name='function'),
                      go.Scatter(x=df['w'][20:80], y=df['target_line'][20:80], name='function'),
                      go.Scatter(x=[w_0], y=[derivative_2(w_0)], name='point', marker_size = 10)],
                layout = go.Layout(width=500, title='Loss function'))
fig.show()
```

```
gradient_descent(derivative = derivative_2, learning_rate = 0.00001, w_0 = -4, max_iters = 1000, precision = 0.00001, epoch = True)
```
